```python
from bs4 import BeautifulSoup
```

BeautifulSoup me permite arreglar todo el texto identarlo y ponerlo un formato legible

```python
s = BeautifulSoup(p12.text, 'lxml')
```

```python
type(s)
```

```python
print(s.prettify())
```

```python
secciones = s.find('ul', attrs={'class':'horizontal-list main-sections hide-on-dropdown'}).find_all('li')
secciones
```

Usando la función find, le pedí que extraiga toda la información relacionada con el atributo UL, adcional a esto le pedí que me extraiga la información relacionada una clase en particular del HTML usando la función attrs y por último le pedi que busque con todos los atributos. Todo esto lo guardé en la variable secciones.

```python
seccion = secciones[0]
seccion
```

creo la variable seccion en donde guardo el primer item de la variable secciones

```python
seccion.find('a')
```

```python
seccion.a.get('href')
```

acá extraigo el link del primer ítem

```python
seccion.a.get_text()
```

acá extraigo el texto del primer ítem

```python
links_secciones = [seccion.a.get('href') for seccion in secciones]
links_secciones 
```

acá cree un for para extraer los href de todas las secciones y las guardé en la variable links_secciones

```python
sec = requests.get(links_secciones[0])
sec
```

```python
sopa_section = BeautifulSoup(sec.text, 'lxml')
```

```python
print(sopa_section.prettify())
```
